```
import numpy as np
python_list = [1, 2, 3, 4, 5]
np_array = np.array(python_list)
print("NumPy Array:", np_array)
print("Array Shape:", np_array.shape)
print("Array Mean:", np_array.mean())
print("Array Sum:", np_array.sum())
```

```
import numpy as np
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([6, 7, 8, 9, 10])
print("Array 1:", array1)
print("Array 2:", array2)
```

```
print("Type of array1:", type(array1))
print("Type of array2:", type(array2))
```

```
multiplication = array1 * array2
division = array1 / array2
power = np.power(array1, array2)
print("Multiplication:", multiplication)
print("Division:", division)
print("Power:", power)
```

```
addition = array1 + array2
print(f"Addition of Two:\n{array1} + {array2} = {addition}")
```

```
sin_values = np.sin(array1)
log_values = np.log(array1 + 1)
log2_values = np.log2(array1 + 1)
exp_values = np.exp(array1)
print("Sin values:", sin_values)
print("Log values:", log_values)
print("Log2 values:", log2_values)
print("Exp values:", exp_values)
```

```
import pandas as pd
python_list = [1, 2, 3, 4, 5]
pd_series = pd.Series(python_list)
print("Pandas Series:\n", pd_series)
print("Series Mean:", pd_series.mean())
print("Series Sum:", pd_series.sum())
print("Series Standard Deviation:", pd_series.std())
```

```
import pandas as pd
import numpy as np
series1 = pd.Series([1, 2, 3, 4, 5])
series2 = pd.Series([6, 7, 8, 9, 10])
print("Series 1:\n", series1)
print("Series 2:\n", series2)
```

```
print("Type of series1:", type(series1))
print("Type of series2:", type(series2))
```

```
multiplication_series = series1 * series2
division_series = series1 / series2
power_series = series1 ** series2
print("Multiplication:\n", multiplication_series)
print("Division:\n", division_series)
print("Power:\n", power_series)
```

```
addition_series = series1 + series2
print(f"Addition of Two:\n{series1.values} + {series2.values} = {addition_series.values}")
```

```
sin_values = np.sin(series1)
log_values = np.log(series1 + 1)
log2_values = np.log2(series1 + 1)
exp_values = np.exp(series1)
print("Sin values:\n", sin_values)
print("Log values:\n", log_values)
print("Log2 values:\n", log2_values)
print("Exp values:\n", exp_values)
```

```
import matplotlib.pyplot as plt
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
```

```
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.scatter(height, weight, color='blue', s=100, alpha=0.7)
plt.title('Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 2)
plt.bar(people, height, color='green', alpha=0.7, label='Height')
plt.title('Height of Individuals')
plt.xlabel('People')
plt.ylabel('Height (cm)')
plt.grid(True)
plt.legend()
plt.show()
plt.figure(figsize=(12, 6))
plt.bar(people, weight, color='red', alpha=0.7, label='Weight')
plt.title('Weight of Individuals')
plt.xlabel('People')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.legend()
plt.show()
```

```
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.hist(age, bins=5, color='purple', edgecolor='black', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.grid(True)
plt.subplot(1, 3, 2)
plt.hist(height, bins=4, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribution of Height')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.subplot(1, 3, 3)
plt.hist(weight, bins=4, color='cyan', edgecolor='black', alpha=0.7)
plt.title('Distribution of Weight')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
